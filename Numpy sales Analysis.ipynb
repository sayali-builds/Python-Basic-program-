```
#Numpy Sales Analysis
import numpy as np

sales=np.array([
12000,15000,18000,17000,
20000,21000,24000,25000,
26000,27000,29000,30000])

print("Total Sales:",np.sum(sales))
print("Average Sales:",np.mean(sales))
print("Highest Sales:",np.max(sales))
print("Lowest Sales:",np.min(sales))

best_month=np.argmax(sales)+1

print("Best Month:",best_month)

growth=((sales[-1]-sales[0])/sales[0])*100
```
