```
#Employee Salary Analysis
import numpy as np

salary=np.array([
25000,30000,35000,
40000,45000,50000,
55000,60000])

print("Mean:",np.mean(salary))
print("Median:",np.median(salary))
print("Std Dev:",np.std(salary))

print("Highest:",np.max(salary))
print("Lowest:",np.min(salary))

print("Range:",np.max(salary)-np.min(salary))
```
